```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tdf = pd.read_csv('Titanic.csv')
```

```python
tdf = pd.DataFrame(tdf)
```

```python
tdf.head()
```

```python
tdf.drop('Cabin', axis=1, inplace= True)
```

```python
tdf.head()
```

```python
tdf.info()
```

```python
tdf.dropna(inplace= True)
```

# PANDAS

```python
survivors_by_sex = tdf.groupby('Sex')['Survived'].sum()
survivors_by_sex.plot.bar(rot=0, legend=1)
```

```python
age_class = tdf.groupby('Pclass')['Age']
age_class.plot.hist(legend=1)
```

```python
survived_class = tdf.groupby('Pclass')['Survived'].sum()
survived_class.plot.pie(legend=1)
```

```python
age_plt = tdf['Age'].dropna().astype(int)
fig = plt.figure()
ax= fig.gca()
ax.set_title('Age')

ax.violinplot(age_plt)
plt.show()
```

```python
class1=[]
class2=[]
class3=[]
for c in tdf['Pclass']:
    if c == 1:
      class1.append(c)
    if c == 2:
      class2.append(c)
    if c == 3:
      class3.append(c)
classes = [len(class1), len(class2), len(class3)]
clabel = ['class 1','class 2','class 3']
plt.pie(classes, labels=clabel, autopct ='%0.1f%%', shadow = True, explode=[0,0.1,0.02])
```

```python
fig = plt.figure()

ax = fig.add_axes([0.1,0.1,0.8,0.8])


ax.plot(tdf['Fare'],tdf['Age'], 'y*')
ax.set_xlabel('Fare')
ax.set_ylabel('Age')
ax.set_title('Fare by Age')

```

# SEABORN

```python
tdf.head()
```

```python
survivors = tdf['Survived'].replace({0:'No', 1:'Yes'})
sns.boxplot(x='Pclass', y='Age', hue = survivors, palette = ['r', 'g'], data= tdf).set(title = 'Surivivors by class')
sns.despine(offset= 20,trim=True)
```

```python
survivors = tdf['Survived'].replace({0:'No', 1:'Yes'})
palette = ['r', 'g']
title = 'Survival by age & class'
sns.violinplot(data=tdf, x= 'Pclass', y = 'Age', hue =survivors, split= True, inner = 'quart', palette = palette, saturation =0.5).set(title = title)
```

```python
title = 'Embarked by Fare and Age'
palette = ['y', 'r', 'b']
sns.lmplot(data = tdf, x= 'Age', y = 'Fare', hue = 'Embarked', palette = palette).set(title = title)
```

```python
title = 'Fare by Class and Sex'
sns.catplot(data = tdf, x = 'Sex', y = 'Fare', hue = 'Pclass', kind = 'boxen', palette = 'colorblind').set(title = title)
```

```python
title = 'Fare by Sex'
sns.scatterplot(data = tdf, x = 'Fare', y = 'Age', size= 'Fare', hue = 'Sex', palette = 'pastel').set(title = title)
```
